## Описание
Лабораторная работа посвящена построению и сравнению моделей множественной линейной регрессии с использованием методов кросс-валидации. В работе реализован полный цикл анализа данных: от предобработки и исследования взаимосвязей до построения прогнозов и интерпретации результатов.

## Основные этапы работы
1. **Подготовка данных** - разделение на обучающую (80%) и тестовую (20%) выборки
2. **Предварительный анализ** - описательные статистики и визуализация взаимосвязей
3. **Проверка нормальности** - анализ распределения целевой переменной Y
4. **Построение моделей** - создание спецификаций для Y и log(Y)
5. **Кросс-валидация** - оценка точности моделей
6. **Прогнозирование** - тестирование лучших моделей на отложенной выборке
7. **Интерпретация** - анализ параметров наиболее точной модели

## Методы
- Множественная линейная регрессия
- Кросс-валидация (метод указан в варианте)
- Оценка качества моделей (MSE)
- Статистический анализ данных

## Требования
- Python 3.x
- Библиотеки: pandas, numpy, scikit-learn, matplotlib, seaborn, scipy
