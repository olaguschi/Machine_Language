## Описание проекта
Этот проект представляет собой реализацию методов снижения размерности и регуляризации логистической регрессии на примере набора данных о качестве красного вина.

## Вариант 9

## Данные
- **Набор данных**: winequalityred_for_lab
- **Зависимая переменная**: Y (бинарная классификация)
- **Объясняющие переменные**:
  - fixed_acidity
  - volatile_acidity
  - citric_acid
  - residual_sugar
  - chlorides
  - free_sulfur_dioxide
  - total_sulfur_dioxide
  - density
  - pH
  - sulphates
  - alcohol

## Методы
- **Метод снижения размерности**: PCR (Principal Component Regression)
- **Метод регуляризации**: Ридж-регрессия (Ridge Regression)

## Структура проекта
1. **Загрузка и предварительный анализ данных**
   - Проверка типов данных
   - Анализ пропущенных значений
   - Описательная статистика
   - Визуализация распределения переменных

2. **Предобработка данных**
   - Стандартизация показателей
   - Разделение на обучающую и тестовую выборки

3. **Методы снижения размерности**
   - Метод главных компонент (PCA)

4. **Регуляризация логистической регрессии**
   - Ридж-регрессия
   - Подбор оптимального параметра регуляризации
   - Визуализация коэффициентов

## Результаты
Проект демонстрирует применение современных методов машинного обучения для:
- Снижения размерности данных
- Регуляризации моделей классификации
- Визуального анализа влияния параметров на модель
